```python
# task 2 if there is a,b,NaN,c,d then running avg is a+b/2 =m that come at the place of nan 
import pandas as pd
import numpy as np

#  for read Project.csv 
proj = pd.read_csv("Project.csv")

# for replacing Nan by running avg
running_sum = 0
count = 0
new_costs = []

for cost in proj["Project Cost"]:
    if pd.notna(cost):
        running_sum += cost
        count += 1
        new_costs.append(cost)
    else:
        avg = running_sum / count if count > 0 else proj["Project Cost"].mean()
        new_costs.append(avg)
        running_sum += avg
        count += 1

proj["Project Cost"] = new_costs
proj.to_csv("Project.csv", index=False)
print(proj)
```

```python
#task 3 col name breaked into two col first name and second name
# to read employee.css we made it in 1st task
emp = pd.read_csv('Employee.csv')

split_names = emp['Name'].str.split(n=1)
emp['First Name'] = split_names.str[0]
emp['Last Name'] = split_names.str[1].fillna('')


# Drop the old Name column
emp = emp.drop(columns=['Name'])


# Save updated Employee CSV (even though later tasks will re-read as required)
emp.to_csv('Employee.csv', index=False)
print(emp)
```

```python
#task 4 merge the three file which we have made at 1st task but now employee.csv is updated in 3rd task it have first name and last name
#project is updated by running avg


emp = pd.read_csv('Employee.csv')
sen = pd.read_csv('Seniority.csv')    # to read csv files
proj = pd.read_csv('Project.csv')


# Merge Employee with Seniority (one-to-one) -> left join on ID
final_df = emp.merge(sen, on='ID', how='left')
# Merge result with Project (one-to-many, because one employee may have many projects)
final_df = final_df.merge(proj, on='ID', how='left')


# Reorder columns for readability
cols = ['ID','First Name','Last Name','Gender','City','Age','Designation Level','Project','Project Cost']
final_df = final_df[cols]


# Save final dataframe to CSV (optional)
final_df.to_csv('Final.csv', index=False)
print('\nTask 4: Created Final dataframe by joining three CSVs and saved as Final.csv')
print(final_df.head(10))
```

```python
#task 5 5%bounus added 
# First, ensure Project Cost numeric

final_df['Project Cost'] = pd.to_numeric(final_df['Project Cost'], errors='coerce')
status_list = ['Finished','Ongoing','Finished','Ongoing','Finished','Failed','Finished','Failed','Ongoing','Finished','Failed','Ongoing','Finished','Finished']
proj['Status'] = status_list
# Save back updated project info
proj.to_csv('Project.csv', index=False)


# Recreate final_df merging again so Status is present
final_df = emp.merge(sen, on='ID', how='left').merge(proj, on='ID', how='left')
final_df = final_df[['ID','First Name','Last Name','Gender','City','Age','Designation Level','Project','Project Cost','Status']]


# Calculate bonus
final_df['Bonus'] = final_df.apply(lambda row: 0.05 * row['Project Cost'] if str(row['Status']).strip().lower() == 'finished' else 0.0, axis=1)


# Save final with bonus
final_df.to_csv('Final.csv', index=False)

print(final_df[['ID','Project','Status','Project Cost','Bonus']])
```

```python
#task 6 give 1 to fail and delete 4 made a temp file final


failed_ids = proj.loc[proj['Status'].str.strip().str.lower() == 'failed', 'ID'].unique()


# Demote designation by 1 for those IDs in the seniority dataframe
sen_updated = sen.copy()
sen_updated.loc[sen_updated['ID'].isin(failed_ids), 'Designation Level'] = sen_updated.loc[sen_updated['ID'].isin(failed_ids), 'Designation Level'] + 1
# Note: The task wording says "Demote the designation level by 1", but designation scale has 1=highest, 4=lowest.
# Demotion increases the numeric level by 1. So we increment the numeric value by 1 for failed employees.


# Delete employees whose designation level is above 4 (i.e., >4)
sen_updated = sen_updated[sen_updated['Designation Level'] <= 4].reset_index(drop=True)


# Now update the Employee and Final dataframes by keeping only IDs present in updated seniority
valid_ids = sen_updated['ID'].unique()
emp_filtered = emp[emp['ID'].isin(valid_ids)].reset_index(drop=True)
final_df = emp_filtered.merge(sen_updated, on='ID', how='left').merge(proj, on='ID', how='left')


print(emp_filtered)
```

```python
#task 7
#Add title based on Gender (M -> Mr., F -> Mrs.)
# Work on final_df which currently has Gender for the filtered employees
final_df['First Name'] = final_df.apply(lambda r: ('Mr. ' if r['Gender'] == 'M' else 'Mrs. ') + r['First Name'], axis=1)


# Drop Gender column
final_df = final_df.drop(columns=['Gender'])


final_df.to_csv('Final.csv', index=False)

print(final_df[['ID','First Name','Last Name']])
```

```python
# task 8

import pandas as pd

# Final.csv read karo
final = pd.read_csv("Final.csv")

# Promotion: Age > 29 ho to Designation Level -1 (minimum 1 rakho)
final['Designation Level'] = final.apply(
    lambda r: max(r['Designation Level'] - 1, 1) if r['Age'] > 29 else r['Designation Level'],
    axis=1
)

# Save wapas Final.csv me
final.to_csv("Final.csv", index=False)

print("Updated Final.csv with promotions:")
print(final[['ID','First Name','Age','Designation Level']])
```
